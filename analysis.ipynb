```python
from models.rnn import RNN
import util
```

```python
data = util.k_fold(4, 'training_data.csv')
val_data = data[0].validation_set
model = RNN()
model.train(val_data)
```

```python
import numpy

hist = model.hist
perplexity = numpy.exp(hist.history['loss'])
accuracy = hist.history['acc']
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.plot(perplexity[2:])
plt.xlabel("Epochs")
plt.ylabel('Perplexity')

plt.subplot(1,2,2)
plt.plot(accuracy)
plt.xlabel("Epochs")
plt.ylabel('Accuracy')
```
